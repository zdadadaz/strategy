```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://isin.twse.com.tw/isin/C_public.jsp?strMode=1'
# url = 'https://www.tej.com.tw/webtej/doc/uid.htm'
r = requests.get(url)
soup = BeautifulSoup(r.content)
```

```python
for a_tag in soup.find_all('tr'):
    for td in a_tag.findAll('td'):
        print(td.text)
    break
```

```python
len(soup.find_all('tr'))
```

```python
out = {'ticker':[],'name':[], 'type':[], 'industry':[]}
for ii,a_tag in enumerate(soup.find_all('tr')):
    if ii == 0 :
        continue
#     print(ii, a_tag)
    for idx,td in enumerate(a_tag.findAll('td')):
        flag = False
        if idx == 0:
            tmp = td.text.split()
            if len(tmp) <2:
                flag = True
                break
            out['ticker'].append(tmp[0])
            out['name'].append(tmp[1])
        elif idx == 3:
            out['industry'].append(td.text)
        elif idx == 4:
            out['type'].append(td.text)
        if flag:
            break
```

```python
out
```

```python
import pandas as pd
out = pd.DataFrame.from_dict(out)
del out['type']
out.to_csv('datas/tw_stocks.csv',index=False, encoding = 'utf_8_sig')
```
